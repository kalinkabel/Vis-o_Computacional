# **Limiarizao Adaptativa**

```python
#funo que retornar as imagens das limiarizaes simples, mdia e gaussiano
def limiarizacao_adaptativa(img, limiar=140,block_size=11, C=4):
  _, limiarizacao_global= cv.threshold(img, limiar,255,cv.THRESH_BINARY)
  limiar_media = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, block_size, C)
  limiar_gauss = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, block_size, C)

  imagens = [img, limiarizacao_global, limiar_media, limiar_gauss]
  titulos = ['Imagem Original','Limiarizao simples', 'Limiarizao mdia', 'Limiarizao gaussiana']

  fig = plt.gcf()
  fig.set_size_inches(18,12)
  
  for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(cv.cvtColor(imagens[i], cv.COLOR_BGR2RGB), cmap='gray')
    plt.title(titulos[i])
    plt.xticks([]),plt.yticks([])
```

```python
limiarizacao_adaptativa(img)
```
